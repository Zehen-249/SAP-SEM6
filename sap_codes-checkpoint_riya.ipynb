```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
M = 10000  # Number of samples
n = 10    # Size of each sample

# Storage for means
meanX = []
meanY = []
meanZ = []
meanC = []

# Generating means of random samples
for _ in range(M):
    X = np.random.binomial(10, 0.2, n)  # Binomial distribution
    Y = np.random.poisson(50, n)        # Poisson distribution
    Z = np.random.normal(50, 0.5, n)   # Normal distribution
    x0 = 0    # Location parameter
    gamma = 1 # Scale parameter
    C = np.random.standard_cauchy(size=n) * gamma + x0
    meanX.append(np.mean(X))
    meanY.append(np.mean(Y))
    meanZ.append(np.mean(Z))
    meanC.append(np.mean(C))


# Plotting the results
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Binomial Distribution
mu_X, sigma_X = np.mean(meanX), np.std(meanX)
axs[0,0].hist(meanX, bins=30, density=True, color='blue', edgecolor='k', label='Binomial Means')
x = np.linspace(min(meanX), max(meanX), 100)
axs[0,0].plot(x, norm.pdf(x, mu_X, sigma_X), 'r--', label='Normal Fit')
axs[0,0].set_title("Binomial Distribution")
axs[0,0].legend()

# Poisson Distribution
mu_Y, sigma_Y = np.mean(meanY), np.std(meanY)
axs[0,1].hist(meanY, bins=30, density=True, color='green', edgecolor='k', label='Poisson Means')
y = np.linspace(min(meanY), max(meanY), 100)
axs[0,1].plot(y, norm.pdf(y, mu_Y, sigma_Y), 'r--', label='Normal Fit')
axs[0,1].set_title("Poisson Distribution")
axs[0,1].legend()

# Normal Distribution
mu_Z, sigma_Z = np.mean(meanZ), np.std(meanZ)
axs[1,0].hist(meanZ, bins=30, density=True, color='orange', edgecolor='k', label='Normal Means')
z = np.linspace(min(meanZ), max(meanZ), 100)
axs[1,0].plot(z, norm.pdf(z, mu_Z, sigma_Z), 'r--', label='Normal Fit')
axs[1,0].set_title("Normal Distribution")
axs[1,0].legend()

# Cauchy-Lorentz Distribution
mu_C, sigma_C = np.mean(meanC), np.std(meanC)
axs[1,1].hist(meanC, bins=30,density=True, color='yellow', edgecolor='k', label='Normal Means')
c = np.linspace(min(meanC), max(meanC), 100)
axs[1,1].plot(c, norm.pdf(c, mu_C, sigma_C), 'r--', label='Normal Fit')
axs[1,1].set_title("Cauchy Lorentz Distribution")
axs[1,1].legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
M = 600  # Number of samples
n = 20    # Size of each sample

# Generate all samples at once
X = np.random.binomial(10, 0.2, (M, n))  # Binomial distribution
Y = np.random.poisson(50, (M, n))        # Poisson distribution
Z = np.random.normal(50, 0.5, (M, n))    # Normal distribution
C = np.random.standard_cauchy((M, n))    # Cauchy-Lorentz distribution

# Compute means along axis 1 (row-wise mean)
meanX = np.mean(X, axis=1)
meanY = np.mean(Y, axis=1)
meanZ = np.mean(Z, axis=1)
meanC = np.mean(C, axis=1)

# Plotting the results
fig, axs = plt.subplots(2, 2, figsize=(8, 8))

# Function to plot histograms with normal fit
def plot_hist(ax, data, color, title):
    mu, sigma = np.mean(data), np.std(data)
    ax.hist(data, bins=30, density=True, color=color, edgecolor='k', label='Sample Means')
    x = np.linspace(min(data), max(data), 100)
    ax.plot(x, norm.pdf(x, mu, sigma), 'r--', label='Normal Fit')
    ax.set_title(title)
    ax.legend()

# Subplots
plot_hist(axs[0,0], meanX, 'red', "MEAN Binomial Distribution")
plot_hist(axs[0,1], meanY, 'green', "MEAN Poisson Distribution")
plot_hist(axs[1,0], meanZ, 'blue', "MEAN Normal Distribution")
plot_hist(axs[1,1], meanC, 'yellow', "MEAN Cauchy Lorentz Distribution")



plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.stats as stats

def coin_toss_simulation(n, q):
    """
    Simulates a coin toss experiment n times with probability q of getting heads.
    """
    return np.random.binomial(1, q, n)  # 1 represents heads, 0 represents tails

def binomial_test(n, q, alpha=0.05):
    """
    Performs a binomial test to check if the coin is fair (H0: q=0.5).
    """
    # Simulate coin tosses
    tosses = coin_toss_simulation(n, q)
    num_heads = np.sum(tosses)  # Count number of heads

    # Perform Binomial Test using binomtest()
    p_value = stats.binomtest(num_heads, n, 0.5, alternative='two-sided').pvalue

    # Print results
    print(f"Number of Tosses (n): {n}")
    print(f"Observed Probability of Heads: {num_heads/n:.4f}")
    print(f"Null Hypothesis: H0 (q = 0.5)")
    print(f"Alternative Hypothesis: H1 (q ≠ 0.5)")
    print(f"P-value: {p_value:.4f}")

    # Conclusion based on significance level alpha
    if p_value < alpha:
        print(f"Reject H0: The coin is biased (Significance level = {alpha})")
    else:
        print(f"Fail to reject H0: No significant evidence of bias")

# Example Usage
n = 100  # Number of tosses
q = 0.5  # Probability of getting heads (change this for testing)
binomial_test(n, q)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

def plot_heatmap(X, Y):
    """
    Automatically generates a heatmap of the joint distribution of two discrete random variables.

    Parameters:
    - X: array-like, first random variable
    - Y: array-like, second random variable
    """
    # Get unique sorted values for X and Y
    x_vals = np.sort(np.unique(X))
    y_vals = np.sort(np.unique(Y))
    
    # Create a mapping from values to indices
    x_map = {val: i for i, val in enumerate(x_vals)}
    y_map = {val: i for i, val in enumerate(y_vals)}
    
    # Create a 2D frequency matrix
    H = np.zeros((len(y_vals), len(x_vals)), dtype=int)
    for x, y in zip(X, Y):
        H[y_map[y], x_map[x]] += 1  # Increment count at the correct index

    # Plot using seaborn heatmap
    plt.figure(figsize=(7, 5))
    sns.heatmap(H.T, annot=True, fmt="d", cmap="viridis",
                xticklabels=x_vals, yticklabels=y_vals)
    
    # Labels and title
    plt.xlabel("X (Random Variable)")
    plt.ylabel("Y (Random Variable)")
    plt.title("Heatmap of Joint Distribution")
    plt.show()

# Example Usage:
np.random.seed(42)  
X = np.random.binomial(n=10, p=0.4, size=10)  # Binomial data
Y = np.random.poisson(lam=4, size=10)  # Poisson data

plot_heatmap(X, Y)  # Just pass the data, no bins needed
print(X)
print(Y)
```

```python
#Unit 2: BINOMIAL HYPOTHESIS TEST NULL HYPOTHESIS

import numpy as np
import random
import scipy.stats as stats
n = 1000 #size
q = 0.5
coin_flips = np.random.choice([0, 1], size=n,p=[1-q, q])  # 0: Tail, 1: Head

num_heads = np.sum(coin_flips)
result = stats.binomtest(num_heads, n, p=0.5, alternative='two-sided')
print(f"Number of Heads: {num_heads}")
print(f"P-value: {result.pvalue}")
if result.pvalue < 0.05:
    print("We reject the null hypothesis: The coin is likely biased.")
else:
    print("We do not have enough evidence to reject the null hypothesis: The coin may be fair.")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
n = 1000  # Number of trials
p = 0.5   # Fair coin probability
num_heads = np.sum(np.random.choice([0, 1], size=n, p=[1-p, p]))  # Observed heads

# Compute binomial distribution
x = np.arange(0, n + 1)
binom_pmf = stats.binom.pmf(x, n, p)

# Binomial test
result = stats.binomtest(num_heads, n, p, alternative='two-sided')
p_value = result.pvalue

# Plot
plt.figure(figsize=(10, 6))
plt.bar(x, binom_pmf, color='blue', alpha=0.6, label="Binomial Distribution")

# Highlight observed value
plt.axvline(num_heads, color='red', linestyle='dashed', label=f'Observed Heads: {num_heads}')

# Mark rejection regions (two-tailed test)
alpha = 0.05  # Significance level
critical_values = [stats.binom.ppf(alpha / 2, n, p), stats.binom.ppf(1 - alpha / 2, n, p)]
plt.axvspan(0, critical_values[0], color='red', alpha=0.3, label="Rejection Region")
plt.axvspan(critical_values[1], n, color='red', alpha=0.3)

# Labels
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p}) with Observed Data")
plt.legend()
plt.show()

print(f"Number of Heads: {num_heads}")
print(f"P-value: {p_value}")
```
